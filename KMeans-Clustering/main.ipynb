```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
df.head()
```

```python
''' Here we do not know what to predict, we only know that
we need to make some categories in K-Means Clustering.
So we do not know what value y will take. '''
```

```python
''' Since we do not know what the output will be, there is no
train test split in clusting. So entier df is used. '''
```

```python
X = df.iloc[:,[3, 4]].values
''' Customer ID is excluded since it is not
important in the making of cluster.
Only anual income and spending score are kept so that
we can easily visualise the clusters'''
''' Since there is no dependant variable pre creation
of the model the "y" value is not taken. '''
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range (1,11):
  km = KMeans(n_clusters=i, init='k-means++', random_state=42)
  km.fit(X)
  wcss.append(km.inertia_)
plt.plot(range(1,11), wcss)
plt.show()
```

```python
''' Value of the dependant variables will be 1,2,3,4,and 5.
These will actually be the 5 groups of clusters.
This will let us classify customers for this dataset into
one of these 5 groups. '''
```

```python
km = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = km.fit_predict(X)
```

```python
print(y_kmeans)
```

```python
''' Visualizing on a 2d plot.
Each cluster is given 1 scatterplot. '''
```

```python
''' Making cluster of index 0. '''
plt.scatter(X[y_kmeans == 0,0],X[y_kmeans == 0,1], s=100, color = 'red')
```

```python
''' Making cluster of index 1. '''
plt.scatter(X[y_kmeans == 1,0],X[y_kmeans == 1,1], s=100, color = 'green')
```

```python
''' Making cluster of index 2. '''
plt.scatter(X[y_kmeans == 2,0],X[y_kmeans == 2,1], s=100, color = 'blue')
```

```python
''' Making cluster of index 3. '''
plt.scatter(X[y_kmeans == 3,0],X[y_kmeans == 3,1], s=100, color = 'purple')
```

```python
''' Making cluster of index 4. '''
plt.scatter(X[y_kmeans == 4,0],X[y_kmeans == 4,1], s=100, color = 'yellow')
```

```python
''' Now plotting the centroid of the 5 groups '''
plt.scatter(km.cluster_centers_[:, 0],km.cluster_centers_[:, 1], s= 300, color = 'pink')
plt.plot()
```

```python
plt.scatter(X[y_kmeans == 0,0],X[y_kmeans == 0,1], s=100, color = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1,0],X[y_kmeans == 1,1], s=100, color = 'green', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2,0],X[y_kmeans == 2,1], s=100, color = 'blue', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3,0],X[y_kmeans == 3,1], s=100, color = 'purple', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4,0],X[y_kmeans == 4,1], s=100, color = 'yellow', label = 'Cluster 5')
plt.scatter(km.cluster_centers_[:, 0],km.cluster_centers_[:, 1], s= 300, color = 'pink')
plt.legend()
plt.xlabel('Annual Income')
plt.ylabel('Spending Score(1-100)')
plt.show()
```
