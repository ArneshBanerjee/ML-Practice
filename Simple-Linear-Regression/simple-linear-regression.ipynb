```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Salary_Data.csv")
X = df.iloc[:,:-1].values #Selects all columns except last
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split #used to import the tt-split 
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0) # splitting into x and y testing and training data
```

```python
from sklearn.linear_model import LinearRegression # linear model is imported from sklearn
lr = LinearRegression() # lr variable is given to LinearRegression Model. The model is only built, it still needs to be fit(trained) on training data
```

```python
# fit method is used to train all models
lr.fit(X_train, y_train) # X_train and y_train are put in to train the lr model on the training data
```

```python
# A model has been made using the .fit(X_train, y_train)
# Now we need to predict the results of the observations in the test set using this model
# The dataset is a years of experience vs salary
# For this dataset we call the precit method on 'Years of Experience' and get predicted 'Salary'
# Then compare the predicted salary to actual salary
```

```python
y_pred = lr.predict(X_test) # X_test is the number of years of experience since we predict salary based on that only
print(y_pred)
```

```python
# Now we need to compare the real salary and predicted salary for training set
plt.scatter(X_train, y_train, color = 'red') # Number of years of exp vs salary
plt.plot(X_train, lr.predict(X_train), color = 'blue') # Years of exp vs predicted salary on training set
plt.title('Salary vs expeprience (Training Set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Plotting actual training set vs predicted training set
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, y_pred, color = 'blue')
plt.title('Salary vs Experience (Test Set)')
```

```python
print(lr.predict([[12]])) # Used to predict one single variable i.e. salary with 12 years of experience/
```

```python
# y = b0 + (b1)x
# to get b0 and b1 where b1 is the coefficient and b0 is the intercept
print(lr.coef_)
print(lr.intercept_)
```
